```python
%%time
# circ = zx.Circuit.load('../circuits/demo/ex2_id_16.qasm')
circ = zx.Circuit.load('../circuits/demo/noend.qasm')
g = circ.to_graph()
g.set_type(3, VertexType.X)
g.set_type(77, VertexType.X)
zx.draw(g, labels=True)
g1 = g.adjoint()
zx.draw(g1, labels=True)
g.compose(g1)
zx.draw(g, labels=True)
g.remove_edge((29, 59))
g.add_vertex(VertexType.X, 3, 12)
g.add_vertex(VertexType.X, 3, 13)
g.add_edge((29, 60))
g.add_edge((59, 61))
zx.draw(g, labels=True)
```

```python
zx.simplify.full_reduce(g, quiet = False)
zx.draw(g, labels=True)
print("#vertices: ", g.num_vertices())
```

```python
print("#vertices: ", g.num_vertices())
zx.draw(g, labels=True)
zx.simplify.spider_simp(g, quiet = False)
print("#vertices: ", g.num_vertices())
zx.draw(g, labels=True)
zx.simplify.id_simp(g, quiet = False)
print("#vertices: ", g.num_vertices())
zx.draw(g, labels=True)
zx.simplify.to_gh(g, quiet = False)
# zx.draw(g, labels=True)
zx.simplify.pivot_simp(g, quiet = False)
print("#vertices: ", g.num_vertices())
zx.draw(g, labels=True)
```
